```python
from tkinter import *
import os
from tkinter import messagebox
import tkinter as tk
import logging
import PyPDF2

def merge():
    """
    To merge all pdf's in a given path to a new pdf doc.
    """
    root = tk.Tk()

    pdfls = []
    w = Label(root, text='Please enter your directory path to see all files', font=('Consolas', 20), width=50)
    w.pack()
    entered_directory_holder = Entry(root, font=('Consolas', 15), width=50)
    entered_directory_holder.pack()
    

    w1 = Label(root, text='Please provide merged pdf file name', font=('Consolas', 20), width=50)
    w1.pack()
    Merging_Filename_holder = Entry(root, font=('Consolas', 15), width=50)
    Merging_Filename_holder.pack()
    
    
    def pdf_merge(pdf_final_file, pdf_dir, pdf_files):

        pdfWriter = PyPDF2.PdfFileWriter()
        for i in pdf_files:
            pdfil = open(pdf_dir + "\\" + i, 'rb')
            pyd = PyPDF2.PdfFileReader(pdfil)

        for pageNum in range(pyd.numPages):
            pageObj = pyd.getPage(pageNum)
            pdfWriter.addPage(pageObj)
            pdfOutputFile = open(pdf_dir + "\\" + pdf_final_file, 'wb')
            pdfWriter.write(pdfOutputFile)
        pdfil.close()

        pdfOutputFile.close()
    
    
    def onclick():
        files = os.listdir(entered_directory_holder.get())
        for f in files:
            cc = ''
            for i in range(len(f) - 4, len(f)):
                cc = cc + f[i]
                if (cc.lower() == '.pdf'):
                    pdfls.append(f)

        

        if (Merging_Filename_holder.get()[-4:].lower() != '.pdf'):
            pdf_merge(str(Merging_Filename_holder.get()) + str('.pdf'), entered_directory_holder.get(), pdfls)
                        
        else:
            pdf_merge(Merging_Filename_holder.get(), entered_directory_holder.get(), pdfls)
            
                
        myButton['state'] = 'disabled'
        
    myButton = Button(root, text="Submit", command=onclick, font=('Consolas', 15))
    myButton.pack()
    root.mainloop()

merge()
```

```python
from tkinter import *
import os

import tkinter as tk

from PyPDF4 import PdfFileMerger


def merge():
    """
    To merge all pdf's in a given path to a new pdf doc.
    """
    root = tk.Tk()

    pdfls = []
    w = Label(root, text='Please enter your directory path to see all files', font=('Consolas', 20), width=50)
    w.pack()
    entered_directory_holder = Entry(root, font=('Consolas', 15), width=50)
    entered_directory_holder.pack()
    

    w1 = Label(root, text='Please provide merged pdf file name', font=('Consolas', 20), width=50)
    w1.pack()
    Merging_Filename_holder = Entry(root, font=('Consolas', 15), width=50)
    Merging_Filename_holder.pack()
    
    
    os.mkdir("merger_processing_folder")
    new = entered_directory_holder.get()+"merger_processing_folder"
    os.chdir(new)
    def MergeButtonCommand():
        merger = PdfFileMerger()
        for j in os.listdir(entered_directory_holder.get()):
            if j.endswith(".pdf"):
                merger.append(open(j, 'rb'))

        
        merger.write(Merging_Filename_holder.get()+".pdf")
        merger.close()


        myButton['state'] = 'disabled'
        
    myButton = Button(root, text="Merge", command=MergeButtonCommand, font=('Consolas', 15))
    myButton.pack()
    root.mainloop()

merge()
```
