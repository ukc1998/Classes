```python
from tkinter import *
import tkinter.filedialog as fd
import tkinter.messagebox as mb
import subprocess, platform
import os
import PyPDF2

file_dic = {}


def list_files_in_folder():
    folder = fd.askdirectory(title='Select the folder whose files you want to list')
    global file_dic

    for root_dir, dirs, files in os.walk(folder):

        for ele in files:
            print(ele)
            file_dic[ele] = root_dir + "/"

    # creating  a new window to display the result
    global list_files_wn
    list_files_wn = Toplevel(root)
    list_files_wn.title('Files in your selected folder')
    list_files_wn.geometry('400x300')
    button_font = ("Times New Roman", 13)
    button_background = 'Green'
    list_files_scrollbar = Scrollbar(list_files_wn, orient=VERTICAL)
    list_files_listbox = Listbox(list_files_wn, width=50, selectbackground='SteelBlue', font=("Georgia", 10),
                                 yscrollcommand=list_files_scrollbar.set, selectmode=MULTIPLE)
    Button(list_files_wn, text='Merge PDF', width=10, font=button_font, bg=button_background,
           command=merge_pdf).place(x=0, y=170)

    # configure scrollbar
    list_files_scrollbar.config(command=list_files_listbox.yview)
    list_files_scrollbar.pack(side=RIGHT, fill=Y)
    list_files_listbox.pack(pady=15)
    # inserting the list into the list
    for i in list(file_dic.keys()):
        list_files_listbox.insert(END, i)


def merge_pdf():
    mergeFile = PyPDF2.PdfFileMerger()
    merge = file_dic.copy()
    count = 0
    for key, values in merge.items():
        if key.endswith("pdf"):
            path = values + key
            mergeFile.append(PyPDF2.PdfFileReader(path, 'rb'))
            count += 1
    if count < 2:
        mb.showerror("No file To merge  ", "Please check file count " + str(count))
    else:
        mergeFile.write("NewMergedFile.pdf")
        mb.showinfo("Merged Done ", "File saved " + str(os.getcwd()) + "Merged PDF count " + str(count))

    list_files_wn.destroy()
    file_to_open = str(os.getcwd()) + "NewMergedFile.pdf"
    if platform.system() == 'Darwin':       # macOS
        subprocess.call(('open', file_to_open))
    elif platform.system() == 'Windows':# Windows
        os.startfile(file_to_open)
    else:                  # linux variants
        subprocess.call(('xdg-open', file_to_open))


def main():

    # Defining the variables
    title = 'File challenge Jan iNeuron '
    background = 'White'
    button_font = ("Times New Roman", 13)
    button_background = 'Green'

    # Initializing the window
    root.title(title)
    root.geometry('400x100')
    root.config(bg=background)

    # Creating and placing the components in the window
    Label(root, bg=background, wraplength=300).place(x=20, y=0)
    Button(root, text='Search files in a folder', width=33, font=button_font, bg=button_background,
           command=list_files_in_folder).place(x=3, y=25)

    # Finalizing the window
    root.update()
    root.mainloop()


if __name__ == "__main__":
    root = Tk()
    main()
```
