```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(20)
df = pd.DataFrame(np.random.randn(10,4),index = pd.date_range("2021-07-18", periods = 10), columns = ["a", "b", "c", "d"])
df
```

```python
df["new"] = df["c"] +df.d
df
```

```python
df["new_col_square_d"] = df.d.apply(lambda x: x**2)
df
```

```python
df["new_col_log_d"] = df.d.apply(lambda x: np.log(x))
df
```

```python
def test(x):
    if x<0:
        return "Negative Number found"
    else:
        return np.log(x)
    
df["new_col_log_d"] = df.d.apply(test)
df
```

```python
def test1(x):
    if x<0:
        return 1
    return 0

df["new_a"] = df["a"].apply(test1)
df
```
