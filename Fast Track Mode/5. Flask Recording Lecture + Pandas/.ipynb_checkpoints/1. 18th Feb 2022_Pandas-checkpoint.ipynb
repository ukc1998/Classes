```python
### second method for the above problem:
user = []
res=requests.get('https://api.github.com/repos/pandas-dev/pandas/issues')
js=res.json()
for i in range(len(js)):
    user.append(js[i]['user'])

pd.DataFrame(user)
```

```python
### third method

pd.DataFrame([js[i]["user"] for i in range(len(js))])
```
