```python
#python code to show box-cox transformation of non-normal data
#import modules

import numpy as np
from scipy import stats

#importing plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

#generating non-normal data i.e. exponential data i.e. Pareto distribution
original_data = np.random.exponential(size=1000)

#transforming training data and saving alfa value
fitted_data, fitted_alpha = stats.boxcox(original_data)

#creating axes to draw plots:
fig, ax = plt.subplots(1,2)


#plotting the original data (non-normal) and transformed data (normal)
sns.distplot(original_data, hist = False, kde = True,
             kde_kws = {'shade' : True, 'linewidth':2},
             label = 'Non-Normal', color = 'green', ax = ax[0])

#plotting the Transformed data (normal) and transformed data (normal)
sns.distplot(fitted_data, hist = False, kde = True,
             kde_kws = {'shade' : True, 'linewidth':5},
             label = 'Normal', color = 'red', ax = ax[1])

#adding legends to the subplots
plt.legend(loc='upper right')

#rescaling the subplots
fig.set_figheight(8)


print(f"Alpha value used for transformation: {fitted_alpha}")
```
