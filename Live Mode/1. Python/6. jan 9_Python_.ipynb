```python
def fib(n):
    a=0
    b=1
    for i in range(n):
        yield a,i
        a,b = b, a+b
```

```python
for i in fib(10):
    print(i)
```

```python
def fib1(n):
    a=0
    b=1
    output = []
    for i in range(n):
        output.append(a)
        a,b = b, a+b
    return output
```

```python
fib(10)
```
